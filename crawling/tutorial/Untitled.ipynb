```python
from bs4 import BeautifulSoup
import requests
```

```python
base_url = 'https://www.rottentomatoes.com/top/bestofrt/'
```

```python
query = {
    'year': '2019'
}
```

```python
url = '{base_url}?{query}'.format(
    base_url=base_url,
    query='&'.join([f'{key}={value}' for key,value in query.items()]))
```

```python
url = 'https://www.rottentomatoes.com/top/bestofrt/?year=2019'
```

```python
req = requests.get(url)
```

```python
req.text
```

```python
if req:
    soup = BeautifulSoup(req.text, 'html.parser')
```

```python
for rank, title in zip(soup.select('table td.bold'), soup.select('table a.articleLink')):
    print(rank.getText().strip(), title.getText().strip(), title.get('href'))
```
