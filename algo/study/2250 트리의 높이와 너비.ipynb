```python
class Node:
    def __init__(self, item, pointer=None):
        self.item = item
        self.pointer = pointer
        self.left = None
        self.right = None
        
    
    def __eq__(self, value):
        return self.item == value


class Tree:
    x_list_per_level = {}
    def __init__(self, node):
        self.root = node
        self.x_count = 0

    def inorder_traversal(self, node, level):
        if node.left is not None:
            self.inorder_traversal(node.left, level + 1 )
        self.x_count += 1
        try:
            self.x_list_per_level[level].append(self.x_count) 
        except KeyError:
            self.x_list_per_level[level] = [self.x_count]
        if node.right is not None:
            self.inorder_traversal(node.right, level + 1)
    

n = int(input())
NodeList = []
for i in range(n):
    n_list = list(map(int, input().split()))
    if i == 0:
        root = Node(n_list[0])
        tree = Tree(root)
        NodeList.append(root)
    else:
        idx = NodeList.index(n_list[0])
        root = NodeList[idx]
    root.left = Node(n_list[1]) if n_list[1] != -1 else None
    root.right = Node(n_list[2]) if n_list[2] != -1 else None
    if n_list[1] != -1 : NodeList.append(root.left)
    if n_list[2] != -1 : NodeList.append(root.right)


tree.inorder_traversal(tree.root, 1)
print()
print(tree.x_list_per_level)

width_per_level = []
for key, value in tree.x_list_per_level.items():
    min_value = min(value)
    max_value = max(value)
    width = max_value - min_value + 1
    width_per_level.append((key, width))
width_per_level.sort(key=lambda x: (x[1], -x[0]), reverse=True)
print(*width_per_level[0])
```
