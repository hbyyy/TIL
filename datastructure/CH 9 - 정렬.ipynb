# 거품 정렬

```python
def bubble_sort(numbers):
    for i in range(len(numbers) -1, 0, -1):
        print(numbers)
        for j in range(i):
            if numbers[j] > numbers[j+1]:
                numbers[j], numbers[j+1] = numbers[j+1], numbers[j]
    
```

# 선택 정렬

```python
def selection_sort(numbers):
    for i in range(len(numbers)-1):
        min_idx = i
        print(numbers)
        for j in range(i+1, len(numbers)):
            if numbers[min_idx] > numbers[j]:
                min_idx = j
        numbers[i], numbers[min_idx] = numbers[min_idx], numbers[i]
```

```python
ll = [11, 3, 28, 43, 9, 4]
```

```python
selection_sort(ll)
```

```python
def insert_sort(numbers):
    for i in range(1, len(numbers)):
        j = i
        while j > 0 and numbers[j-1] > numbers[j]:
                numbers[j-1], numbers[j] = numbers[j], numbers[j-1]
                j -= 1
```

```python
def count_sort(numbers):
    count = {}
    count_arr = [0 for i in range(max(numbers) +1)]
    result = [0 for _ in range(len(numbers))]
    
    for number in numbers:
        try:
            count[number].append(number)
        except KeyError:
            count[number] = [number]
            
    for key in count.keys():
        count_arr[key] = len(count[key])
    
    for i in range(1, len(count_arr)):
        count_arr[i] = count_arr[i] + count_arr[i-1]
        
    for i in range(len(numbers)-1, -1, -1):
        count_arr[numbers[i]] -= 1
        result[count_arr[numbers[i]]] = numbers[i]
    
    return result
```

```python
numbers = list()
for _ in range(int(input())):
    numbers.append(int(input()))
    
def quick_sort(seq):
    if len(seq) < 2:
        return seq
    idx = len(seq) // 2
    pivot = seq[idx]
    
    before = list() 
    after = list()
    
    for i in range(len(seq)):
        if i == idx:
            continue
        if seq[i] <= pivot:
            before.append(seq[i])
        else:
            after.append(seq[i])
    
    return quick_sort(before) + [pivot] + quick_sort(after)

for num in quick_sort(numbers):
    print(num)
```
